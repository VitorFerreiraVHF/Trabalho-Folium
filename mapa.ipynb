```python
import folium
import pandas as pd
from folium.plugins import HeatMap

url_geojson = "https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson"

data = {
    "id": ["AC", "AL", "AM", "AP", "BA", "CE", "DF", "ES", "GO", "MA", "MT", "MS",
           "MG", "PA", "PB", "PE", "PI", "PR", "RJ", "RN", "RO", "RR", "RS", "SC", "SE", "SP", "TO"],
    "populacao": [881935, 3337357, 4144597, 845731, 14873064, 9132078, 3015268, 4018650,
                  7113540, 7075181, 3484466, 2778986, 21168791, 8602865, 4039277, 9557071,
                  3273227, 11433957, 17264943, 3506853, 1777225, 605761, 11377239, 7164788,
                  2298696, 45919049, 1572866],
    "area": [164123, 27778, 1570745, 142815, 564733, 148920, 5802, 46095, 340086, 331983,
             903357, 357145, 586528, 1247954, 56585, 98311, 251577, 199307, 43780, 52811,
             237590, 224299, 281748, 95736, 21915, 248222, 277620]
}

df = pd.DataFrame(data)
df["densidade"] = df["populacao"] / df["area"]

centroides = {
    "AC": [-9.97499, -67.8243], "AL": [-9.5713, -36.782], "AM": [-3.4653, -62.2159],
    "AP": [1.4142, -51.775], "BA": [-12.9704, -38.5124], "CE": [-3.7172, -38.5433],
    "DF": [-15.8267, -47.9218], "ES": [-19.9191, -40.3374], "GO": [-16.6869, -49.2643],
    "MA": [-2.5387, -44.2829], "MT": [-15.6009, -56.0979], "MS": [-20.4428, -54.6465],
    "MG": [-19.9167, -43.9345], "PA": [-1.4558, -48.5044], "PB": [-7.2399, -35.8808],
    "PE": [-8.0476, -34.877], "PI": [-5.0892, -42.8016], "PR": [-25.4296, -49.2713],
    "RJ": [-22.9068, -43.1729], "RN": [-5.7945, -35.211], "RO": [-8.7612, -63.9004],
    "RR": [2.8194, -60.6732], "RS": [-30.0346, -51.2177], "SC": [-27.5954, -48.548],
    "SE": [-10.9472, -37.0731], "SP": [-23.5505, -46.6333], "TO": [-10.1842, -48.3331]
}

heat_data = [
    [centroides[estado][0], centroides[estado][1], densidade]
    for estado, densidade in zip(df["id"], df["densidade"])
]

m = folium.Map(location=[-15.788497, -47.879873], zoom_start=4)
HeatMap(heat_data, radius=25, max_zoom=10).add_to(m)
m.save("heatmap_brasil.html")
m
```
